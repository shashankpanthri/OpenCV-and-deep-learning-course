# Morphological Operators 

morphological operators are sets of kernels that can achieve variety of effects like reducing noise

certain operators are very good at reducing black points on a white background and vice versa

certain opertors can also achieve an erosion or dilation effect that can add or erode from an existing image

This effect can most easily be seen on text so we will work some of these operators on white text on black background

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

### first the erosion operator

what this does is erode the boundaries of the foreground in the image,

in the above image the letters are foreground and the black at back is background

Works best when foreground is light color (preferrably white) and background is dark.

```python
kernel = np.ones(shape=(5,5), dtype=np.uint8)
```

```python
erode = cv2.erode(img,kernel,iterations=1)
display_img(erode)
```

now the more iterations the image goes through the more eroded the foreground text will become

```python
erode = cv2.erode(img,kernel,iterations=3)
display_img(erode)
```

## Opening
Opening is erosion followed by dilation. Useful in removing background noise!

```python
img = load_img()
```

first lets just add some background noise to the image first

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600)) # this will create the array of same size as the original image with
                                                            # all vaules either one or zero
```

```python
white_noise
```

```python
img.max()
```

since max value of the img is 255 we have to make the noise also take max value as 255

```python
white_noise = white_noise * 255
```

```python
white_noise
```

now we can add this noise to the image

```python
noise_img = img+white_noise
```

```python
display_img(noise_img)
```

now we can use opening to clear the background noise from the image

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
```

```python
display_img(opening)
```

it may not be perfect as we can see there are some distortions in the image but coming from the above image, it is pretty darn good


# Closing
 Useful in removing noise from foreground objects, such as black dots on top of the white text.

```python
img = load_img()
```

we will now add some foreground noise to the text

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
black_noise
```

```python
black_noise= black_noise * -255
black_noise
```

```python
black_noise_img = img + black_noise
black_noise_img
```

```python
black_noise_img[black_noise_img== -255] = 0
black_noise_img
```

```python
black_noise_img.min()
```

```python
display_img(black_noise_img)
```

now we can use the closing to reduce the foreground noise

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_img(closing)
```

agian the image is not the best but considering where it came from it is still pretty good

## Morphological Gradient

Difference between dilation and erosion of an image.

```python
img = load_img()
```

```python
display_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```

 we can use this one as a not very good edge detection methods
 we will learn more about other edge detection methods that are better than this
